```python
import pandas as pd
import sqlite3
import altair as alt
```

```python
con = sqlite3.connect("games.db")
df = pd.read_sql_query("SELECT * from games", con)
```

```python
print(df.head())
```

```python
alt.Chart(df).encode(
        x='game_id',
        y='dice_throws'
    ).mark_line()
```

```python
alt.Chart(df).encode(
        x='game_id',
        y='player_turns'
    ).mark_line()
```
